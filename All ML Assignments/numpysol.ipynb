```python
# Q2 
# Create a NumPy array of shape (4, 4) with values from 1 to 16. Replace the diagonal elements with 0.
arr = np.arange(1,17).reshape(4,4)
for i in range(len(arr)):
    arr[i][i]=0
print(arr)
```

```python
# Q2
# Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border.
arr2= np.random.randint(1,25,size =(5,5))
a = arr2[0,:]
b = arr2[-1,:]
c = arr2[:,0]
d = arr2[:,-1]
print(np.concatenate((a,b,c,d) ,axis=0))

```

```python
# Assignment 3
# Q1
# Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.
a = np.random.randint(1,10,size=(3,4))
b= np.random.randint(10,20,size=(3,4))
print("Array a:")
print(a)
print("Array b:")
print(b)
print("Element-wise addition:")
print(a+b)
print("Element-wise subtraction:")
print(a-b)
print("Element-wise multiplication:")
print(a*b)
print("Element-wise division:")
print(a/b)

```

```python
# Q2
# Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.
arr3 = np.arange(1,17).reshape(4,4)
print("Array:")
print(arr3)
print("Row-wise sum:")
print(np.sum(arr3, axis=1))

print("Column-wise sum:")
print(np.sum(arr3, axis=0))
```

```python
# Assignment 4
# Q1
# Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.
e = np.random.randint(1,25,size =(5,5))
print("Array:")
print(e)
print(np.mean(e))
print(np.median(e))
print(np.std(e))
```

```python
#Q2
# Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).
arr4 = np.arange(1,10).reshape(3,3)
normalized = (arr4 - np.mean(arr4))/np.std(arr4)
print(normalized)
```

```python
# Assignment 5
# Q1
# Create a NumPy array of shape (3, 3) filled with random integers. Add a 1D array of shape (3,) to each row of the 2D array using broadcasting.
m = np.random.randint(1,10,size=(3,3))
n = np.random.randint(1,10,size=(3,))
print(m + n)
```

```python
# Assignment 6
# Q1
# 1. Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.
s = np.random.randint(1,10,size=(3,3))
print("Matrix:")
print(s)
print("Determinant:")
print(np.linalg.det(s))
print("Inverse:")
print(np.linalg.inv(s))
print("Eigenvalues:")
print(np.linalg.eigvals(s)) 
```

```python
# Q2
# Create two NumPy arrays of shape (2, 3) and (3, 2). Perform matrix multiplication on these arrays.
s = np.random.randint(1,10,size=(2,3))
t = np.random.randint(1,10,size=(3,2))
print("Matrix s:")
print(s)
print("Matrix t:")
print(t)
print("Matrix multiplication of s and t:")
print(np.dot(s,t))
```
